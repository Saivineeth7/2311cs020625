```python
import pandas as pd
from google.colab import files

uploaded = files.upload()
file_name = list(uploaded.keys())[0]
df = pd.read_csv(file_name)

# Now use df
df.head()

# Assuming your movie title column is named 'movie'e.
user_movies_df = df.pivot(index='userId', columns='movie', values='rating')
# If the movie title column has a different name, replace 'movie' with the actual nam
```

```python
user_movies_df.head()
```

```python
user_movies_df.index = df.userId.unique() # Use df instead of movies_df
user_movies_df.head()
```

```python
user_movies_df.fillna(0, inplace=True)
user_movies_df.head()
```

```python
user_sim_df = pd.DataFrame(index=user_movies_df.index, columns=user_movies_df.index)
user_sim_df.head()
```

```python
!pip install scikit-learn  # Install scikit-learn if you haven't already
from sklearn.metrics.pairwise import pairwise_distances # Import the necessary function

user_sim = 1 - pairwise_distances(user_movies_df.values, metric='cosine')
user_sim
```
